```python
# LIBRERÍAS
!uv add -U ultralytics

from ultralytics import YOLO
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt
import cv2
```

## Prueba de predicción

```python
# Ruta a los pesos entrenados
weights = "../data/runs/detect/train_results/weights/best.pt"

# Cargar el modelo
model = YOLO(weights)

# Ruta a la imagen o carpeta que quieres analizar
image_path = "../data/images/test/test_000012.jpg" 
```

```python
# Realizar predicción
results = model.predict(source=image_path, conf=0.25, save=True, save_txt=True, )

# Mostrar los resultados en la primera imagen procesada
img_path = results[0].path 
img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
